```python
import seaborn as sns
```

```python
#Loading data file & defining df
df = pd.read_csv('Cancer_data.csv')
```

```python
#Generating statistics
perimeter_mean_stats = df['perimeter_mean'].describe()
print(perimeter_mean_stats)
```

```python
#Visualizing the distribution
plt.figure(figsize=(18, 10))

# Histogram
sns.histplot(df['perimeter_mean'], kde=True)
plt.title('Distribution of Perimeter Mean')
plt.xlabel('Perimeter Mean')
plt.ylabel('Frequency')
plt.show()

# Boxplot
plt.figure(figsize=(18, 10))
sns.boxplot(x=df['perimeter_mean'])
plt.title('Boxplot of Perimeter Mean')
plt.xlabel('Perimeter Mean')
plt.show()
```

```python
#Identifying outliers using Z scores
df['z_score'] = stats.zscore(df['perimeter_mean'])
outliers = df[(df['z_score'] > 3) | (df['z_score'] < -3)]
print(outliers)
```

```python
#Visualization of Marginal distribution
plt.figure(figsize=(18, 10))
sns.histplot(df['perimeter_mean'], kde=True)
plt.title('Marginal Distribution of Perimeter Mean')
plt.xlabel('Perimeter Mean')
plt.ylabel('Probability Density')
plt.show()
```

```python
df = pd.read_csv('Cancer_data.csv')
```

```python
#Finding the probability of a tumor
# Count the number of tumors with perimeter_mean <= 20
count_less_equal_20 = df[df['perimeter_mean'] <= 20].shape[0]

# Total number of tumors
total_count = df.shape[0]

# Probability calculation
probability = count_less_equal_20 / total_count
print(f"Probability of a tumor having a perimeter_mean is: {probability:.4f}")
```
